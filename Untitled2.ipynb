```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline


datao = pd.read_csv('datao.csv', sep=',')
datap = pd.read_csv('data_pure.csv', sep=',')
```

```python
import re

def rep1(x):
    if(x.group()=='1'):
        return '0'  
    elif(x.group()=='0'):
        return '1'
    elif(x.group()=='2'):
        return '2'
    
def rep02(x):
    if(x.group()=='1'):
        return '2'  
    elif(x.group()=='2'):
        return '1'
    elif(x.group()=='0'):
        return '0'

def rep2(x):
    if(x.group()=='2'):
        return '0'  
    elif(x.group()=='0'):
        return '1'
    elif(x.group()=='1'):
        return '2'
    

boardsi = datap['bo'].to_numpy()

boards = datap['bo'].to_numpy().copy()
boardsr = np.ones((len(boards), 8), dtype=int)

for j, board in enumerate(boards):
    board = str(board)
    board_parts = re.split("(\w{3})", board)[1::2]
    
    for i, bp in enumerate(board_parts):
        if(i in [0, 4]):
            bpn = re.sub("1|0|2", rep1, bp)
            board_parts[i] = bpn
        elif(i in [1, 5]):
            bpn = bp
            board_parts[i] = bpn
        elif(i in [2, 6]):
            bpn = re.sub("1|0|2", rep02, bp)
            board_parts[i] = bpn
        elif(i in [3, 7]):
            bpn = re.sub("1|0|2", rep2, bp)
            board_parts[i] = bpn
            
    boardsr[j]=board_parts

# a = ['000', \'001', '002', '010', '011', '012', '020', '021', '022', '100', '101', '102', '110', '111', '112', '120', '121', '122', '200', '201', '202', '210', '211', '212', '220', '221', '222']
# a = ['000', \'001', '010', '011', '020', '021', '022', '100', '101', '102', '110', '111', '112', '120', '121', '122', '200', '201', '202', '210', '211', '212', '220', '221', '222']

# I know we have 27 figures we need to represent but if want to take advantage of order we must think again. I want to 
# align the order with the progress of the game. Small number will indicate the the beginning of the game while 
# big number indicate more moves had been taken.

# This 27 : 0 is for the for the initial line for line one is '111' , line 2 and 3 '000 ' , line 4 '222'.
# The 7 initial numbers [1 to 7] for the line to be occupied by its initial number and the most favorite number 
# to occupy: example for line 1 the favorite is 0 so the combination of 1 and  0 is 8 (extracting '000')
# so '001', '010', '011', '100', '101', '110' '111' given [1 to 7]
# when the two numbers are mixed we get [6] figures: [8-13]
# when occupied by the intial and the unfavorite [0] and [2] excluding '222' given [14-19]
# when occupied by the favorite and the unfavorite [1] and [2] excluding '222' given [20-25]
# '222' is given 26


"""

so lets make it more compact: 

'000' given 0
mix of 0 and 1 excluding '000' given [1-7]. match : '0|1'
mix of the three excluding '000', '111', '222' given [8-13]
mix of 0 and 2 excluding '000' and '222' given [14-19]
mix of 1 and 2 excluding '111' and '222' given [20-25]
'222' given 26

"""

a = np.ones(4)
def nrep(x):
    if(x in [10, 100, 101, 11, 110, 111]):
        a = [10, 100, 101, 11, 110, 111]
        i = a.index(x)
        x = 2 + i
        
    elif(x in [12, 21, 102, 120, 201, 210]):
        a = [12, 21, 102, 120, 201, 210]
        i = a.index(x)
        x = 8 + i
    elif(x in [2, 20, 200, 202, 22, 220]):
        a = [2, 20, 200, 202, 22, 220]
        i = a.index(x)
        x = 14 + i
    elif(x in [112, 121, 211, 122, 212, 221]):
        a = [112, 121, 211, 122, 212, 221]
        i = a.index(x)
        x = 20 + i
    elif(x == 222):
        x = 26
    return x

vnrep = np.vectorize(nrep)

d = vnrep(boardsr)

df = pd.DataFrame(d, columns=['1l1', '1l2', '1l3', '1l4', '2l1', '2l2', '2l3', '2l4'])
df = df.join(datap['win'])

# sns.relplot(x='1l1', y='win', data=df)
# df['win'].plot.hist()
df.loc[:, df.loc[:,'1l1'] == 23] 
```
